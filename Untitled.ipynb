# Loadning the book

```python
with open("miracle_in_the_andes.txt","r",encoding="utf8") as file:
    book=file.read()
```

## The most used words(non articles)

```python
import re
pattern=re.compile("[a-zA-Z]+")
findings=re.findall(pattern,book.lower())
findings[:5]
```

```python
d = {}
for word in findings:
    if word in d.keys():
        d[word] = d[word] + 1
    else:
        d[word] = 1
```

```python
d_list = [(value,key) for (key,value) in d.items()]
d_list = sorted(d_list,reverse=True)
d_list[:5]
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
english_stopwords = stopwords.words("english")
```

```python
filtered_words = []
for count, word in d_list:
    if word not in english_stopwords:
        filtered_words.append((word,count))
```

## Sentiment Analysis: what is the most positive and most negative chapter?

#### An exmaple

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
scores = analyzer.polarity_scores("Hey, look how bad the trees are. I hate them. I really hate them.")
```

```python
if scores["pos"] > scores["neg"]:
    print("This is a positive text")
else:
    print("It is negative text")
```

```python
scores = analyzer.polarity_scores(book)
```

```python
if scores["pos"] > scores["neg"]:
    print("This is a positive book")
else:
    print("It is negative book")
```

## Chapter Sentiment Analysis

```python
import re
patter = re.compile("Chapter [0-9]+")
chapters = re.split(pattern,book)
```

```python
chapters = chapters[1:]
```

```python
for chapter in chapters:
    scores = analyzer.polarity_scores(chapter)
    print(scores)
```
